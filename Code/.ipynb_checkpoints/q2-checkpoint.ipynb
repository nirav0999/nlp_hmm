```python
def readTrainingSet(filepath = "../Dataset/train.np"):
    rows = []
    with open(filepath) as newfile:
        rs = newfile.readlines()
        for r in rs:
            r = r.strip("\n")
            new_r = r.split("\t")
            new_r[0] = new_r[0].lower()
            rows.append(new_r)
            #print(new_r)
    #rows = openCSVfile(filepath,delim = "\t")
    #print("R = ",rows)
    data = []
    cp = 0
    arr = [["<startWord>","<startTag1>","<startTag2>"]]
    while(cp < len(rows)):
        if len(rows[cp]) != 1:
            arr.append(rows[cp])
        else:
            end = ["<endWord>","<endTag1>","<endTag2>"]
            arr.append(end)
            data.append(arr)
            #print(arr)
            arr = [["<startWord>","<startTag1>","<startTag2>"]]
        cp += 1  
    #print(data)
    return data
```

```python
def insertFeatureCountinDict(dictionary,word,tag = None,levels = 1):
    if levels == 1:
        if word not in dictionary:
            dictionary[word] = 1
        else:
            c = dictionary[word] + 1
            dictionary[word] = c
        return dictionary
    else:
        if word not in dictionary:
            dictionary[word] = {tag : 1}
        else:
            if tag not in dictionary[word]:
                dictionary[word][tag] = 1
            else:
                c = dictionary[word][tag] + 1
                dictionary[word][tag] = c
    return dictionary


def calculateFeatures(rows):
    startFeature = {}
    endFeature = {}
    wordPrevPOSTag = {}
    wordNextPOSTag = {}
    wordBIOTag = {}
    for sentence in rows:
        firstWord = sentence[1][0]
        lastWord = -1
        #-------------------------
        for word_no in range(len(sentence) - 1):
            if sentence[word_no + 1][0] == ".":
                lastWord = sentence[word_no][0]
        #-------------------------
        startFeature = insertFeatureCountinDict(startFeature,firstWord,levels = 1)
        endFeature = insertFeatureCountinDict(endFeature,lastWord,levels = 1)
        #----------------------------------------
        for word_no in range(1,len(sentence) - 1):
            word = sentence[word_no][0]
            prevPOSTag = sentence[word_no - 1][1] 
            nextPOSTag = sentence[word_no + 1][1]
            BIOTag = sentence[word_no][2]
            if sentence[word_no + 1][0] == ".":
                nextPOSTag = "<endTag1>"
                break
            wordPrevPOSTag = insertFeatureCountinDict(wordPrevPOSTag,word,prevPOSTag,levels = 2)
            wordNextPOSTag = insertFeatureCountinDict(wordNextPOSTag,word,nextPOSTag,levels = 2)
            wordBIOTag = insertFeatureCountinDict(wordBIOTag,word,BIOTag,levels = 2)
    return startFeature,endFeature,wordPrevPOSTag,wordNextPOSTag,wordBIOTag

trainData = readTrainingSet()
startFeature,endFeature,wordPrevPOSTag,wordNextPOSTag,wordBIOTag = calculateFeatures(trainData)
#print(startFeature)
#print(endFeature)
#print(wordPrevPOSTag)
#print(wordNextPOSTag)
#print(wordBIOTag)
```
